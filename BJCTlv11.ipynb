```python
### N개의 수가 주어졌을 때, 이를 오름차순으로 정렬하는 프로그램을 작성하시오.
N = int(input())
alist = []
alist.append(0)
alist.append(1001)
for i in range(0,N):
    a = int(input())
    for j in range(0,len(alist)-1):
        if alist[j] < a and alist[j+1] > a:
            alist.insert(j+1,a)
for i in range(1,len(alist)-1):
    print(alist[i])
    
    
N = int(input())
alist = []
for i in range(0,N):
    a = int(input())
    alist.append(a)
alist.sort()
for i in range(0,N):
    print(alist[i])
```
