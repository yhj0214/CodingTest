```python
N, M = map(int, input().split())
alist = list(map(int, input().split()))
blist = [0]
clist = []
total = 0
for i in range(0,len(alist)):
    total = total + alist[i]
    blist.append(total)
for i in range(0, M):
    a, b = map(int, input().split())
    clist.append(blist[b]-blist[a-1])
for i in range(0,len(clist)):
    print(clist[i])
```

```python
N, K = map(int, input().split())
alist = list(map(int, input().split()))
max = 0
for i in range(0,K):
    max = max + alist[i]
    temp = max
    
for i in range(0,N-K):
    temp = temp+alist[i+K]-alist[i]
    if temp > max:
        max = temp
print(max)
```

```python
N, M = map(int, input().split())
alist = list(map(int, input().split()))
count = 0
for K in range(1,len(alist)+1):
    temp = 0
    for i in range(0,K):
        temp = temp + alist[i]
    if temp % M == 0:
        count+=1
        
    for i in range(0,N-K):
        temp = temp+alist[i+K]-alist[i]
        if temp % M ==0:
            count = count+1
        
print(count)
```
