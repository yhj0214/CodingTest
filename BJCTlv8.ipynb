```python
# 자연수 M과 N이 주어질 때 M이상 N이하의 자연수 중 소수인 것을 모두 골라 이들 소수의 합과 최솟값을 찾는 프로그램을 작성하시오.

# 예를 들어 M=60, N=100인 경우 60이상 100이하의 자연수 중 소수는 61, 67, 71, 73, 79, 83, 89, 97 총 8개가 있으므로, 이들 소수의 합은 620이고, 최솟값은 61이 된다.
def sosu(num):
    temp = 0
    if num == 1:
        return 1
    else:
        for i in range(2,num):
            if num % i == 0:
                temp = 1
        if temp == 1:
            return 1#소수가 아님
        elif temp == 0:
            return 0 #소수임
            
M = int(input())
N = int(input())
alist = []
for i in range(M, N+1):
    if sosu(i) == 0:
        alist.append(i)


total = 0
if len(alist) == 0:
    print(-1)
else:
    for i in range(0,len(alist)):
        total = total + alist[i]
    print(total)
    print(alist[0])
```

```python
#정수 N이 주어졌을 때, 소인수분해하는 프로그램을 작성하시오.
N = int(input())
alist = []
t = 2

while N > 1:
    if N % t == 0:
        alist.append(t)
        N = N/t
    else:
        t = t+1
for i in range(0,len(alist)):
    print(alist[i])
```

```python
#M이상 N이하의 소수를 모두 출력하는 프로그램을 작성하시오.
M, N = map(int, input().split())
alist = [False for i in range(N+1)]
for i in range(2,N):
    if alist[i] == False:
        j = i
        while j* i < N+1:
            alist[j*i] = True
            j = j+1

for i in range(2, N+1):
    if alist[i] == False and i >= M:
        print(i)
```

```python
# 베르트랑 공준은 임의의 자연수 n에 대하여, n보다 크고, 2n보다 작거나 같은 소수는 적어도 하나 존재한다는 내용을 담고 있다.

# 이 명제는 조제프 베르트랑이 1845년에 추측했고, 파프누티 체비쇼프가 1850년에 증명했다.

# 예를 들어, 10보다 크고, 20보다 작거나 같은 소수는 4개가 있다. (11, 13, 17, 19) 또, 14보다 크고, 28보다 작거나 같은 소수는 3개가 있다. (17,19, 23)

# 자연수 n이 주어졌을 때, n보다 크고, 2n보다 작거나 같은 소수의 개수를 구하는 프로그램을 작성하시오.

def countsosu(M, N):
    total = 0
    alist = [False for i in range(N+1)]
    for i in range(2,N):
        if alist[i] == False:
            j = i
            while j* i < N+1:
                alist[j*i] = True
                j = j+1

    for i in range(2, N+1):
        if alist[i] == False and i > M:
            total = total + 1
    print(total)
countlist = []


while True:
    n = int(input())
    if n == 0:
        break
    else:
        countlist.append(n)
for i in range(0, len(countlist)):
    countsosu(countlist[i],countlist[i]*2)
```

```python
N = 10000
sosulist = [False for i in range(N+1)]
sosu = []
for i in range(2,N):
    if sosulist[i] == False:
        j = i
        sosu.append(i)
        while j* i < N+1:
            sosulist[j*i] = True
            j = j+1

num = int(input())
numlist = []
for i in range(0, num):
    q = int(input())
    numlist.append(q)

for i in range(0, len(numlist)):
    k = 0
    while(sosu[k]<numlist[i]/2):
        k = k+1
        if numlist[i] - sosu[k] in sosu:
            max = sosu[k]
            min = numlist[i] - sosu[k]
    print(min, max)
```

```python
# 한수는 지금 (x, y)에 있다. 직사각형은 각 변이 좌표축에 평행하고, 왼쪽 아래 꼭짓점은 (0, 0), 오른쪽 위 꼭짓점은 (w, h)에 있다. 직사각형의 경계선까지 가는 거리의 최솟값을 구하는 프로그램을 작성하시오.
x ,y, w, h = map(int, input().split())
ylength = h - y
ylength2 = y
xlength = w - x
xlength2 = x
miny = minx = 0
if ylength2 > ylength:
    miny = ylength
else:
    miny = ylength2
if xlength2 > xlength:
    minx = xlength
else:
    minx = xlength2

if minx > miny:
    mini = miny
else:
    mini = minx
print(mini)
```
